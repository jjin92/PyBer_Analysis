```python
# imports
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# using the yahoo finance API to get 30d historical data on spy
spdr = yf.Ticker("SPY")

hist = spdr.history(period="30d")
```

```python
hist.head()
```

```python
# loading covid data
data = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
data.head()
```

```python
# extract unique values from iso_code column
data.iso_code.unique()
```

```python
data.shape
```

```python
# Filter the data for only USA
us_data = data[data["iso_code"]=="USA"]
```

```python
# Set the index of both dfs (DataFrames) to their respective date columns, coverting them to datetime format in the process 
us_data.index = pd.to_datetime(us_data.date)
hist.index = pd.to_datetime(hist.index)
```

```python
# merging the two dfs with a left join on the hist df. This means we will keep only the values in the left df 
# while joining columns for those dates from the right df. 
combined_data = pd.merge(left=hist,right=us_data, how="left",left_index=True, right_index=True)
```

```python
# result is a dataframe whiich contains 30 dates which is what we had in our hist dataframe
combined_data.shape
```

```python
combined_data.head()
```

```python
# identify columns of interest which we will need later for plotting
combined_data.columns
```

```python
# check to see what the index of our new df is
combined_data.index
```

Next, we are going to create a dual axis plot to evaluate SPY vs. New COVID cases

```python
# define our figure size
plt.figure(figsize=(10,6))

# plot the "Close" columns 
combined_data.Close.plot()

# label the left x axis and style it
plt.ylabel("$SPY", color="blue", fontsize=16)

# plot the "new_cases" column with the argument "secondary_y=True" which is necessary here to enforce a dual axis
combined_data.new_cases.plot(secondary_y=True)

# label the right x axis
plt.ylabel("New Covid Cases", color="Orange", fontsize=16)

# give our plot a title
plt.title("$SPY vs. New Covid-19 Cases (30d)", fontsize=20)
```
